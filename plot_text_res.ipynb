```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(10, 81, 10)  # X-axis: # of Tokens
y1 = np.array([0.0733, 0.1215, 0.1610, 0.1908, 0.2151, 0.2320, 0.2470, 0.2597])  # Ours

y2 = np.array([0.0642, 0.1119, 0.1484, 0.1764, 0.1992, 0.2189, 0.2358, 0.2503]) # AttLRP
y3 = np.array([0.0326, 0.0591, 0.0824, 0.1022, 0.1195, 0.1326, 0.1436, 0.1535]) # GradCAM
y4 = np.array([0.0234,  0.0448, 0.0621, 0.0766,0.0892, 0.1026, 0.1136, 0.1234]) # Random
y5 = np.array([0.0258,  0.0474, 0.0653, 0.0815,0.0956, 0.1100, 0.1217, 0.1320]) # Rise


# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='Ours', color='red')
plt.plot(x, y2, label='AttLRP', color='magenta')
plt.plot(x, y3, label='GradCAM', color='green')
plt.plot(x, y4, label='Random', color='cyan')
plt.plot(x, y5, label='RISE', color='blue')
# plt.plot(x, y6, label='rollout', color='yellow')

# Labels and legend
plt.xlabel('# of Tokens')
plt.ylabel('token-F1')
plt.legend(loc='best')

# Show plot
plt.show()
```
